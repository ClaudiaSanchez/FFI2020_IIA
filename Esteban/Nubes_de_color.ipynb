```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import cv2
import os
import glob
from PIL import Image, ImageDraw, ImageFilter, ImageColor
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
import math
```

```python
Car1 = cv2.imread("20210906_S01_2_c.PNG")
Car1 = cv2.cvtColor(Car1, cv2.COLOR_BGR2RGB) 
Car2 = cv2.imread("20210908_S01_2_c.PNG")
Car2 = cv2.cvtColor(Car2, cv2.COLOR_BGR2RGB) 
Car3 = cv2.imread("20210910_S01_2_c.PNG")
Car3 = cv2.cvtColor(Car3, cv2.COLOR_BGR2RGB) 
plt.figure()
plt.subplot(1,3,1)
plt.imshow(Car1)
plt.subplot(1,3,2)
plt.imshow(Car2)
plt.subplot(1,3,3)
plt.imshow(Car3)
plt.show()
```

```python
def RecorrerColor(h,Pix_rec):
  h = h + Pix_rec
  idx = np.where(h > 180)[0]
  h[idx] = h[idx] - 180
  return h
```

```python
plt.figure()
plt.subplot(1,3,1)
plt.imshow(Car1)
plt.subplot(1,3,2)
plt.imshow(Car2)
plt.subplot(1,3,3)
plt.imshow(Car3)
plt.show()


nrows,ncols,nch = Car1.shape
hsvCar1 = cv2.cvtColor(Car1, cv2.COLOR_RGB2HSV)
labCar1 = cv2.cvtColor(Car1, cv2.COLOR_RGB2LAB)
Ximg = np.reshape(Car1,(nrows*ncols,3))
hsvXimg = np.reshape(hsvCar1,(nrows*ncols,3))
labXimg = np.reshape(labCar1,(nrows*ncols,3))
idx = np.logical_and.reduce((Ximg[:,0]!=0 ,Ximg[:,1]!=0,Ximg[:,2]!=0))
idx = np.where(idx)
reds = Ximg[idx]
redshsv = hsvXimg[idx]
redlab = labXimg[idx]

idx = np.random.permutation(len(redshsv))[:600]
colorslab = redlab[idx,:]
colorshsv = redshsv[idx,:]
colors = reds[idx,:]

H = colorshsv[:,0]
H = RecorrerColor(H, 20)
S = colorshsv[:,1]
V = colorshsv[:,2]

R = colors[:,0]
G = colors[:,1]
B = colors[:,2]

C = colorslab[:,0]
I = colorslab[:,1]
E = colorslab[:,2]

fig = plt.figure()
fig.suptitle('Carne 1 HSV')
ax = fig.gca(projection='3d')
ax.scatter(H, S, V, c=colors/255)
ax.set_xlim(0, 180)
ax.set_ylim(0, 255)
ax.set_zlim(0, 255)
ax.set_xlabel('HUE')
ax.set_ylabel('SATURATION')
ax.set_zlabel('VALUE')
plt.show()

fig = plt.figure()
fig.suptitle('Carne 1 RGB')
ax = fig.gca(projection='3d')
ax.scatter(R, G, B, c=colors/255)
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_zlim(0, 255)
ax.set_xlabel('RED')
ax.set_ylabel('GREEN')
ax.set_zlabel('BLUE')
plt.show()

fig = plt.figure()
fig.suptitle('Carne 1 CIELab')
ax = fig.gca(projection='3d')
ax.scatter(C, I, E, c=colors/255)
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_zlim(0, 255)
ax.set_xlabel('C')
ax.set_ylabel('I')
ax.set_zlabel('E')
plt.show()
# _____________________________________________________________Segunda carne_____________________________________________________________
print("_____________________________________________________________Segunda carne_____________________________________________________________")
hsvCar2 = cv2.cvtColor(Car2, cv2.COLOR_RGB2HSV)
labCar2 = cv2.cvtColor(Car2, cv2.COLOR_RGB2LAB)
Ximg = np.reshape(Car2,(nrows*ncols,3))
hsvXimg = np.reshape(hsvCar2,(nrows*ncols,3))
labXimg = np.reshape(labCar2,(nrows*ncols,3))
idx = np.logical_and.reduce((Ximg[:,0]!=0 ,Ximg[:,1]!=0,Ximg[:,2]!=0))
idx = np.where(idx)
reds = Ximg[idx]
redshsv = hsvXimg[idx]
redlab = labXimg[idx]

idx = np.random.permutation(len(redshsv))[:600]
colorslab = redlab[idx,:]
colorshsv = redshsv[idx,:]
colors = reds[idx,:]

H = colorshsv[:,0]
H = RecorrerColor(H, 20)
S = colorshsv[:,1]
V = colorshsv[:,2]

R = colors[:,0]
G = colors[:,1]
B = colors[:,2]

C = colorslab[:,0]
I = colorslab[:,1]
E = colorslab[:,2]

fig = plt.figure()
fig.suptitle('Carne 2 HSV')
ax = fig.gca(projection='3d')
ax.scatter(H, S, V, c=colors/255)
ax.set_xlim(0, 180)
ax.set_ylim(0, 255)
ax.set_zlim(0, 255)
ax.set_xlabel('HUE')
ax.set_ylabel('SATURATION')
ax.set_zlabel('VALUE')
plt.show()

fig = plt.figure()
fig.suptitle('Carne 2 RGB')
ax = fig.gca(projection='3d')
ax.scatter(R, G, B, c=colors/255)
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_zlim(0, 255)
ax.set_xlabel('RED')
ax.set_ylabel('GREEN')
ax.set_zlabel('BLUE')
plt.show()

fig = plt.figure()
fig.suptitle('Carne 2 CIELab')
ax = fig.gca(projection='3d')
ax.scatter(C, I, E, c=colors/255)
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_zlim(0, 255)
ax.set_xlabel('C')
ax.set_ylabel('I')
ax.set_zlabel('E')
plt.show()
# _____________________________________________________________Tercera carne_____________________________________________________________
print("_____________________________________________________________Tercera carne_____________________________________________________________")
hsvCar3 = cv2.cvtColor(Car3, cv2.COLOR_RGB2HSV)
labCar3 = cv2.cvtColor(Car3, cv2.COLOR_RGB2LAB)
Ximg = np.reshape(Car3,(nrows*ncols,3))
hsvXimg = np.reshape(hsvCar3,(nrows*ncols,3))
labXimg = np.reshape(labCar3,(nrows*ncols,3))
idx = np.logical_and.reduce((Ximg[:,0]!=0 ,Ximg[:,1]!=0,Ximg[:,2]!=0))
idx = np.where(idx)
reds = Ximg[idx]
redshsv = hsvXimg[idx]
redlab = labXimg[idx]

idx = np.random.permutation(len(redshsv))[:600]
colorslab = redlab[idx,:]
colorshsv = redshsv[idx,:]
colors = reds[idx,:]

H = colorshsv[:,0]
H = RecorrerColor(H, 20)
S = colorshsv[:,1]
V = colorshsv[:,2]

R = colors[:,0]
G = colors[:,1]
B = colors[:,2]

C = colorslab[:,0]
I = colorslab[:,1]
E = colorslab[:,2]

fig = plt.figure()
fig.suptitle('Carne 3 HSV')
ax = fig.gca(projection='3d')
ax.scatter(H, S, V, c=colors/255)
ax.set_xlim(0, 180)
ax.set_ylim(0, 255)
ax.set_zlim(0, 255)
ax.set_xlabel('HUE')
ax.set_ylabel('SATURATION')
ax.set_zlabel('VALUE')
plt.show()

fig = plt.figure()
fig.suptitle('Carne 3 RGB')
ax = fig.gca(projection='3d')
ax.scatter(R, G, B, c=colors/255)
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_zlim(0, 255)
ax.set_xlabel('RED')
ax.set_ylabel('GREEN')
ax.set_zlabel('BLUE')
plt.show()

fig = plt.figure()
fig.suptitle('Carne 3 CIELab')
ax = fig.gca(projection='3d')
ax.scatter(C, I, E, c=colors/255)
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_zlim(0, 255)
ax.set_xlabel('C')
ax.set_ylabel('I')
ax.set_zlabel('E')
plt.show()
```
